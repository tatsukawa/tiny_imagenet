```python
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as ptick 
import matplotlib.cm as cm
import pandas as pd
from IPython.display import display
```

```python
root_path = os.path.join('..', 'logs')

train_batch_loss_path = os.path.join(root_path, 'run_batch_size-500-tag-Train batch loss.csv')
val_acc_path = os.path.join(root_path, 'run_batch_size-500-tag-Validation accuracy.csv')
val_loss_path = os.path.join(root_path, 'run_batch_size-500-tag-Validation loss.csv')
```

```python
data = pd.read_csv(train_batch_loss_path)
epochs = (data.Step+1)/200
plt.figure(figsize=(16, 8))
plt.rcParams['font.size'] = 26
plt.plot(epochs, data.Value, linewidth=1)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.grid(which='both')
plt.savefig('train_loss.png')
```

```python
data = pd.read_csv(val_acc_path)
epochs = (data.Step+1)/200
plt.figure(figsize=(16, 8))
plt.rcParams['font.size'] = 26
plt.plot(epochs, data.Value, '-o', color='r')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.grid(which='both')
plt.savefig('../images/val_acc.png')
```
